# Objective:

In this analysis focuses on evaluating the impact of the COVID-19 pandemic on India's employment landscape. The dataset being examined includes detail about unemployment rates rates in the different Indian states.

Dataset Details:

The dataset provides insights into the unemployment scenario across different Indian states:

States: The states within India.
Date: The date when the unemployment rate was recorded.
Measuring Frequency: The frequency at which measurements were taken (Monthly).
Estimated Unemployment Rate (%): The percentage of individuals unemployed in each state of India.
Estimated Employed Individuals: The count of people currently employed.
Estimated Labour Participation Rate (%): The proportion of the working population (age group: 16-64 years) participating in the labor force, either employed or actively seeking employment.

This dataset helps in understanding the fluctuation of unemployment across various states in India amid the COVID-19 crisis. It provides valuable insights into the influence of the pandemic on factors such as unemployment rate, employment numbers, and labor participation rates in different regions of the country. The analysis aims to illuminate the socio-economic ramifications of the pandemic on India's workforce and labor market.

# Formatting the columns and their data types-

```python
# Renaming the columns for a better understanding -
df.columns = ['States', 'Date', 'Frequency', 'Estimated Unemployment Rate', 'Estimated Employed',
              'Estimated Labour Participation Rate', 'Region', 'longitude', 'latitude']

df.head(2)
```

```python
# Converting 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], dayfirst = True)
```

# By setting dayfirst=True, we are instructing the function to interpret the date string with the understanding that the day part comes before the month.

```python
# Converting 'Frequency' and 'Region' columns to categorical data type -
df['Frequency'] = df['Frequency'].astype('category')
df['Region'] = df['Region'].astype('category')

# Extracting month from 'Date' and creating a 'Month' column -
df['Month'] = df['Date'].dt.month

# Converting 'Month' to integer format -
df['Month int'] = df['Month'].apply(lambda month: int(month))

# Mapping integer month values to abbreviated month names -
df['Month name'] = df['Month int'].apply(lambda x: calendar.month_abbr[x])

# Dropping the original 'Month' column -
df.drop(columns = 'Month', inplace = True)
```

```python
# Statistical analysis grouped by different regions -
region_stats = df.groupby(['Region'])[['Estimated Unemployment Rate', 'Estimated Employed',
                                       'Estimated Labour Participation Rate']].mean().reset_index()

region_stats
```

# **Creating a Heat Map**

```python
# Selecting specific columns from the DataFrame
map = df[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'longitude', 'latitude', 'Month int']]

# Calculating the correlation matrix for the selected columns
map = map.corr()

# Creating a figure for the heatmap with a specific size
plt.figure(figsize=(6, 4))

# Setting the context for the seaborn plot, adjusting the font scale
sns.set_context('paper', font_scale=1)

# Creating a heatmap using seaborn, with annotations and a specific color map
sns.heatmap(data=map, annot=True, cmap='cividis')
```

## Box plot : Unemployment rate per states

```python
fig = px.box(df, x = 'States',
             y = 'Estimated Unemployment Rate',
             color = 'States',
             title = 'Unemployment rate per states',
             template = 'seaborn')

# Updating the x-axis category order to be in descending order
fig.update_layout(xaxis = {'categoryorder' : 'total descending'})
fig.show()
```

## Bar-plot : Average Unemployment Rate in each state

```python
# Selecting and grouping the data
unemployed = df[['Estimated Unemployment Rate', 'States']]
df_unemployed = unemployed.groupby('States').mean().reset_index()

# Sorting the data by the average unemployment rate
df_unemployed = df_unemployed.sort_values('Estimated Unemployment Rate', ascending=False)

# Creating the bar plot
fig = px.bar(df_unemployed, x='States', y='Estimated Unemployment Rate', color='States',
             title='Average Unemployment Rate in Each State',
             template='plotly',
             labels={'Estimated Unemployment Rate': 'Average Unemployment Rate'})

# Updating the x-axis category order to be in descending order
fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.show()
```

Observation : Haryana, Tripura, Jharkhand are the top unemployed states

## Scatter Matrix : Estimated Unemployment Rate, Estimated Employed, Estimated Labour Participation Rate

```python
# Selecting the relevant columns
scatter_data = df[['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate', 'Region']]

# Creating the scatter matrix plot
fig = px.scatter_matrix(scatter_data,
                        dimensions=['Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate'],
                        color='Region',
                        template='seaborn',
                        title='Scatter Matrix of Unemployment Rate, Estimated Employed, and Labour Participation Rates',
                        labels={'Estimated Unemployment Rate': 'Unemployment Rate',
                                'Estimated Employed': 'Employed',
                                'Estimated Labour Participation Rate': 'Labour Participation Rate'})
fig.show()
```

# Bar-plot(Animation) : Unemployment Rate Across Regions (Jan. 2020 to Oct. 2020)

```python
# Creating the bar plot with animation
fig = px.bar(df,
             x='Region',
             y='Estimated Unemployment Rate',
             animation_frame='Month name',
             color='States',
             title='Unemployment Rate Across Regions (Jan. 2020 to Oct. 2020)',
             height=600,
             template='seaborn')

# Updating the x-axis category order to be in descending order
fig.update_layout(xaxis={'categoryorder': 'total descending'})

# Adjusting the animation frame duration
fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 2000      #2000 = 2000 miliseconds(2 sec)

fig.show()
```

It can be observed that the 'North' region experienced the highest unemployment impact, with the 'South' region following closely across the months from January to October 2020.

# Dashboard analyzing the unemployment rate of each Indian state

```python
# Selecting relevant columns
unemployed_df = df[['States', 'Region', 'Estimated Unemployment Rate', 'Estimated Employed', 'Estimated Labour Participation Rate']]

# Grouping by Region and States, calculating the mean, and resetting the index
unemployed = unemployed_df.groupby(['Region', 'States'])['Estimated Unemployment Rate'].mean().reset_index()

# Creating a Sunburst chart
fig = px.sunburst(unemployed,
                  path=['Region', 'States'],
                  values='Estimated Unemployment Rate',
                  color_continuous_scale='viridis',
                  title='Average Unemployment Rate in Each Region and State',
                  height=550,
                  template='presentation')
fig.show()
```

# Analysis by Scatter Geo plot :

```python
fig = px.scatter_geo(df,'longitude', 'latitude',
                     color="Region",
                     hover_name="States",
                     size="Estimated Unemployment Rate",
                     animation_frame="Month name",
                     scope='asia',
                     template='seaborn',
                     title='Impact of lockdown on Employment across regions')

fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 2000

fig.update_geos(lataxis_range=[5,35], lonaxis_range=[65, 100],oceancolor="#3399FF",
    showocean=True)

fig.show()
```
